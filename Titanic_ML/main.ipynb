## Feature Engineering
1. Categorical Feature Encoding
2. Normalization 

### Categorical Feature Encoding
1. Ordinal Encoding -> Ordinal feature 변환할 때 (순서 있는 것들) e.g) 학력, 선호도 ...
2. One-Hot Encoding -> Nominal feature 변환할 때  e.g) 성별, 부서, 출신학교..

```python
# Categorical Feature Encoding
train_OHE = pd.get_dummies(train, columns=['Sex', 'Embarked'])
train_OHE
```

### Nomalization
모든 데이터가 같은 중요도로 판단되게 하기 위해 scaling을 해준다

```python
# Nomalization  -> Min-Max Scaling

X = train_OHE.drop(columns='Survived')  # input matrix
y = train_OHE.Survived                             # target vector

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
# scaler.fit()
# scaler.transform()
# fit과 transform을 한번에
temp = scaler.fit_transform(X.loc[: ,['Age','Fare']])
X.Age = temp[:, 0]
X.Fare = temp[:, 1]
X
```

# Training

```python
from sklearn.linear_model import SGDClassifier               # 1. Linear Classifier # Stocastic Gradient Descent
from sklearn.linear_model import LogisticRegression      # 2. Logistic Regression
from sklearn.tree import DecisionTreeClassifier              # 3. Decision Tree (CART)
from sklearn.ensemble import RandomForestClassifier  # 4. Random Forest

# 평가 지표
from sklearn.metrics import accuracy_score
```

```python
# 모델 선언
clf = SGDClassifier()
clf2 = LogisticRegression()
clf3 = DecisionTreeClassifier()
clf4 = RandomForestClassifier()

# 학습
clf.fit(X, y)
clf2.fit(X, y)
clf3.fit(X, y)
clf4.fit(X, y)

# 예측  # 학습 결과의 성능 보기
pred = clf.predict(X)
pred2 = clf2.predict(X)
pred3 = clf3.predict(X)
pred4 = clf4.predict(X)
```

```python
print('1. Linear Classifier, Accuracy for Training : %.4f' % accuracy_score(y,pred))
print('2. Logistic Regression, Accuracy for Training : %.4f' % accuracy_score(y,pred2))
print('3. DecisionTree Classifier, Accuracy for Training : %.4f' % accuracy_score(y,pred3))
print('4. RandomForest Classifier, Accuracy for Training : %.4f' % accuracy_score(y,pred4))
```
