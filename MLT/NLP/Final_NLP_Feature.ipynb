```python
import pandas as pd
import numpy as np
import random
```

```python
news_feature = pd.read_csv('news_feature.csv').drop(columns=['Unnamed: 0'])
c38_feature = pd.read_csv('38_label_o.csv').drop(columns=['content','title'])
```

```python
news_feature['date'] = news_feature['date'].str.replace('.','-')
```

```python
c38_feature
```

## 날짜를 어떻게 합해야 할까..

교집합

```python
merge_inner = pd.merge(news_feature, c38_feature)
merge_inner
```

합집합

```python
merge_outer = pd.merge(news_feature, c38_feature, how='outer', on='date')
merge_outer = merge_outer.sort_values(['date']).reset_index().drop(columns=['index'])
```

```python
merge_outer = merge_outer.rename(columns={'label_x' : 'news', 'label_y' : '38'})
merge_outer
```

```python
merge_outer.info()
```

```python
min_date = merge_outer[merge_outer['date'] < '2021-09-27'].index
merge_outer = merge_outer.drop(min_date)
merge_outer
```

```python
merge_outer = merge_outer.fillna(-999)
```

```python
merge_outer.info()
```

```python
merge_outer['news'] = merge_outer['news'].astype(int)
merge_outer['38'] = merge_outer['38'].astype(int)
```

```python
merge_outer
```

```python
merge_outer[merge_outer['date'] == '2020-09-04']
```

```python
merge_date = list(np.array(merge_outer.date.tolist()))
merge_label_news = list(np.array(merge_outer.news.tolist()))
merge_label_38 = list(np.array(merge_outer['38'].tolist()))
```

```python
merge_label_38
```

```python
merge_outer_news = pd.Series(list(np.array(merge_outer.news)))
merge_outer_news_encoding = pd.get_dummies(merge_outer_news)

merge_outer_38 = pd.Series(list(np.array(merge_outer['38'])))
merge_outer_38_encoding = pd.get_dummies(merge_outer_38)
```

```python
merge_outer_38_encoding
```

```python

df = pd.DataFrame({'date':merge_date,
                   'Community_N': list(merge_outer_38_encoding[0]),
                   'Community_P':list(merge_outer_38_encoding[1]),
                   'Community_Nan':list(merge_outer_38_encoding[-999]),
                   'News_N':list(merge_outer_news_encoding[0]),
                   'News_P':list(merge_outer_news_encoding[1]),
                   'News_Nan':list(merge_outer_news_encoding[-999]),
                  })
df
```

```python
df.to_csv('NLP_feature.csv')
```
