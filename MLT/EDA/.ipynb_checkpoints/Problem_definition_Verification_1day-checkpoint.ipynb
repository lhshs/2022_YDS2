### cj 일자별 변화율

```python
cj_rate = []
kospi_rate = []

for i in range(5620):
    cj_rate.append((cj_scaledd.Close[i+1]-cj_scaledd.Close[i])/cj_scaledd.Close[i] * 100)
    kospi_rate.append((kospi_scaledd.Close[i+1]-kospi_scaledd.Close[i])/kospi_scaledd.Close[i] * 100)
    
```

```python
cj_rate.insert(0, 0)
cj_rate
```

```python
kospi_rate.insert(0, 0)
kospi_rate
```

```python
kospi_scaledd['index']
```

```python
df = pd.DataFrame({'date':kospi_scaledd['date'],
                   'Kospi_Rate':kospi_rate,
                   'CJ_Rate':cj_rate})
df
```

```python
test = df[:100]
```

```python
kospi_low = df[(df['Kospi_Rate'] < 0) & (df['Kospi_Rate'] > df['CJ_Rate'])]
kospi_low['day_low'] = -1
kospi_low
```

```python
kospi_scaledd = kospi_scaledd.drop(columns=['level_0'])
```

```python
kospi_scaledd.rename(columns={'index':'date'}, inplace=True)
```

```python
kospi_low
```

```python
kospi_loww = pd.merge(kospi_scaledd, kospi_low, on='date', how='left')
kospi_loww = kospi_loww.fillna(0)
kospi_loww.day_low = kospi_loww.day_low.replace(-1, -10)
kospi_loww
```

```python
kospi_loww = kospi_loww[:100]
```

```python
# cj 대한통운 가격과 kospi 지수 변동율 그래프

plt.figure(figsize=(20, 10))
plt.rcParams['font.family'] = 'AppleGothic'
ax = sns.lineplot(data=df, x="date", y="Kospi_Rate", label='Kospi_Rate', color='blue')
ax = sns.lineplot(data=df, x="date", y="CJ_Rate", label='CJ_Rate', color='orange')
ax = sns.scatterplot(data=kospi_loww, x="date", y="day_low", label='↓', color='black')

ax.set_xlabel('Date')
ax.set_ylabel('Rate')

plt.show()
```
