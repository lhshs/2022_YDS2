## Feature Selection

- 코스피일일등락률 - 주가일일등락률 > 0 인 날을 주가가 하락세인 부분으로 설정하여 '주가하락세'라는 y값을 설정
- 설정한 y값에 대한 xgboost로 5일단위 시계열 예측을 해서 Feature Importance가 낮은 feature를 제거하려고 함

```python
feature_test = pd.read_csv('data/4_Y_value_define.csv')
feature_test.date = pd.to_datetime(feature_test.date)
feature_test = feature_test.fillna(0)
feature_test.head()
# feature_test.isnull().sum()
```

```python
# 주가하락세를 만드는데 쓰인 열들 제거
feature_test = feature_test.drop(['price', '전일 주가', '일일등락률', 'kospi지수', 'kospi등락률(%)', 'kospi대비 주가', 
                                  '현재지수', '대비', '등락률(%)', '시가지수', '고가지수', '저가지수'], axis=1)
feature_test = feature_test.set_index('date')

# 컬럼값을 lag하는 함수 정의
def lag_features(df:pd.DataFrame, col_list:list, lag_num:int)->pd.DataFrame:
  for col in col_list:
    for i in range(1,lag_num+1):
        df[col+'_d+'+str(i)] = df[col].shift(i).fillna(method = 'ffill').fillna(method = 'bfill')
  return df

# lag features 
col_list = list(feature_test.iloc[:, 1:].columns)
feature_test = lag_features(feature_test, col_list, 5)
feature_test.drop(feature_test.index[:5])

feature_test = feature_test.reset_index()

feature_test
```

```python
# SEED 설정 
SEED = 42

# NMAE 함수 정의
def NMAE(true, pred):
 return np.mean(abs(true - pred) / true) * 100 

# public 
xgb_oof_score = []

# split train, valid, test (public)
train = feature_test[:-15]
test = feature_test[-15:-10]

X_train = train[:-5].drop(columns = ['date', '주가하락세'])
y_train = train[:-5]['주가하락세']
X_valid = train[-5:].drop(columns = ['date', '주가하락세'])
y_valid = train[-5:]['주가하락세']
X_test = test.drop(columns = ['date', '주가하락세'])  

# xgboost 
xgb = XGBRegressor(n_estimators= 10000, learning_rate = 0.05, random_state = SEED)
xgb.fit(X_train, y_train, eval_set = [(X_valid, y_valid)], early_stopping_rounds = 100, eval_metric = 'mae', verbose = 100)

# xgb out of fold
xgb_oof = xgb.predict(X_valid)
xgb_pred = xgb.predict(X_test)


# validation NMAE score 
nmae = NMAE(y_valid, xgb_oof)
print(f"\n{train[-5:]['date'].min().strftime('%Y-%m-%d')} ~ {train[-5:]['date'].max().strftime('%Y-%m-%d')} 기간동안의 Public OOF NMAE Score : {nmae:.6f}\n")
xgb_oof_score.append(nmae)
print('=' * 120)

print(f'전체 종목에 대한 평균 Public OOF NMAE Score : {np.mean(xgb_oof_score)}')

# feature importance 확인
plt.rcParams['font.family'] = 'AppleGothic'
fig, ax = plt.subplots(figsize=(20,20))

xgboost.plot_importance(xgb, title='feature_importances', xlabel='', grid=False, ax=ax)
# xgboost.plot_importance(xgb, title='feature_importances', xlabel='', grid=False, ax=ax)
# xgboost.plot_importance(xgb, importance_type='gain', title='gain', xlabel='', grid=False, ax=ax)
# xgboost.plot_importance(xgb, importance_type='cover', title='cover', xlabel='', grid=False, ax=ax)
# xgboost.plot_importance(xgb, importance_type='weight', title='weight', xlabel='', grid=False, ax=ax)
# xgboost.plot_importance(xgb, importance_type='total_gain', title='total_gain', xlabel='', grid=False, ax=ax)
plt.tight_layout()
plt.show()
```

```python
from scipy import interpolate

feature_test = pd.read_csv('data/5_monthly_rawdata.csv')
feature_test.date = pd.to_datetime(feature_test.date)
feature_test.iloc[:,-4:] = feature_test.iloc[:,-4:].fillna(0)
feature_test.head()

# 빈 날짜들 채워 넣어줌
feature_test['base'] = 0
date_series = feature_test.groupby(["base"])['date']
date_series = date_series.apply(lambda x:pd.date_range(start=x.min(), end=x.max(), freq="D")).explode().reset_index()
feature_test = date_series.merge(feature_test, how='left')
feature_test = feature_test.drop(['base'], axis=1)

# 월별 데이터만 추출
month_data = feature_test.dropna()
values = []

for i in range(10,99):
    k = month_data.iloc[:,i].values
    values.append(k)

month_values = pd.DataFrame(values)

#9월 데이터 추가
month_values[12] = np.array(feature_test.iloc[339,10:99])
month_values.iloc[:38,12] = month_values.iloc[:38,11] 
month_values

# Interpolation으로 사이값 채워넣기
# 21.9 ~22.8 각 월 1일의 Index 번호
list_x_0_37 = [0, 4, 35, 65, 96, 127, 155, 186, 216, 247, 277, 308, 369]
list_x_38_88 = [0, 4, 35, 65, 96, 127, 155, 186, 216, 247, 277, 308, 339, 369]

for i in range(38):
    x = np.array(list_x_0_37)
    y = month_values.iloc[i,:]
    f_quadr = interpolate.interp1d(x, y, kind='quadratic')
    y_new_quadr = f_quadr(range(369))
    feature_test.iloc[:,i+10] = y_new_quadr

for i in range(38,89):
    x = np.array(list_x_38_88)
    y = month_values.iloc[i,:]
    y[13] = y[12]
    f_quadr = interpolate.interp1d(x, y, kind='quadratic')
    y_new_quadr = f_quadr(range(369))
    feature_test.iloc[:,i+10] = y_new_quadr

feature_test = feature_test.ffill()
print(feature_test.isnull().sum())
feature_test.head(10)
```

```python
# 종가 데이터 시계열 분해
feature_test = feature_test.set_index('date')
decomposition = seasonal_decompose(feature_test[:-5]['price'], model='multiplicative') 
feature_test['price_trend'] = decomposition.trend.fillna(method = 'ffill').fillna(method = 'bfill') # 종가 추세 
feature_test['price_seosonal'] = decomposition.seasonal.fillna(method = 'ffill').fillna(method = 'bfill') # 종가 계절성 

# 이동평균 추가하는 함수 정의 
def add_rolling_mean(df:pd.DataFrame, col_name:str, window_list:list)->pd.DataFrame:
  for window in window_list:
    df[col_name + f'_rolling{window}'] = df[col_name].rolling(window).mean().fillna(method = 'bfill') 
  return df

# 5일, 20일, 60 이동평균 추가
feature_test = add_rolling_mean(feature_test, 'price', [5, 20, 60]) 

# 여기까지 추가한 feature 저장
feature_test = feature_test.reset_index()
feature_test.to_csv('6_feature_engineering.csv', index=False)
feature_test = feature_test.set_index('date')

# 주가하락세를 만드는데 쓰인 열들 제거
feature_test = feature_test.drop(['price', '전일 주가', '일일등락률', 'kospi지수', 'kospi등락률(%)', 'kospi대비 주가', 
                                  '현재지수', '대비', '등락률(%)', '시가지수', '고가지수', '저가지수'], axis=1)

# 컬럼값을 lag하는 함수 정의
def lag_features(df:pd.DataFrame, col_list:list, lag_num:int)->pd.DataFrame:
  for col in col_list:
    for i in range(1,lag_num+1):
        df[col+'_d+'+str(i)] = df[col].shift(i).fillna(method = 'ffill').fillna(method = 'bfill')
  return df

# lag features 
col_list = list(feature_test.iloc[:, 1:].columns)
feature_test = lag_features(feature_test, col_list, 5)
feature_test.drop(feature_test.index[:5])

feature_test = feature_test.reset_index()

feature_test
```

```python
# SEED 설정 
SEED = 42

# NMAE 함수 정의
def NMAE(true, pred):
 return np.mean(abs(true - pred) / true) * 100 

# public 
xgb_oof_score = []

# split train, valid, test (public)
train = feature_test[:-15]
test = feature_test[-15:-10]

X_train = train[:-5].drop(columns = ['date', '주가하락세'])
y_train = train[:-5]['주가하락세']
X_valid = train[-5:].drop(columns = ['date', '주가하락세'])
y_valid = train[-5:]['주가하락세']
X_test = test.drop(columns = ['date', '주가하락세'])  

# xgboost 
xgb = XGBRegressor(n_estimators= 10000, learning_rate = 0.05, random_state = SEED)
xgb.fit(X_train, y_train, eval_set = [(X_valid, y_valid)], early_stopping_rounds = 100, eval_metric = 'mae', verbose = 100)

# xgb out of fold
xgb_oof = xgb.predict(X_valid)
xgb_pred = xgb.predict(X_test)


# validation NMAE score 
nmae = NMAE(y_valid, xgb_oof)
print(f"\n{train[-5:]['date'].min().strftime('%Y-%m-%d')} ~ {train[-5:]['date'].max().strftime('%Y-%m-%d')} 기간동안의 Public OOF NMAE Score : {nmae:.6f}\n")
xgb_oof_score.append(nmae)
print('=' * 120)

print(f'전체 종목에 대한 평균 Public OOF NMAE Score : {np.mean(xgb_oof_score)}')

# feature importance 확인
plt.rcParams['font.family'] = 'AppleGothic'
fig, ax = plt.subplots(figsize=(20,20))

xgboost.plot_importance(xgb, title='feature_importances', xlabel='', grid=False, ax=ax)
# xgboost.plot_importance(xgb, title='feature_importances', xlabel='', grid=False, ax=ax)
# xgboost.plot_importance(xgb, importance_type='gain', title='gain', xlabel='', grid=False, ax=ax)
# xgboost.plot_importance(xgb, importance_type='cover', title='cover', xlabel='', grid=False, ax=ax)
# xgboost.plot_importance(xgb, importance_type='weight', title='weight', xlabel='', grid=False, ax=ax)
# xgboost.plot_importance(xgb, importance_type='total_gain', title='total_gain', xlabel='', grid=False, ax=ax)
plt.tight_layout()
plt.show()
```

## Modeling - LSTM, Sequential

```python
# 이동평균선까지 feature가 생성된 데이터 불러오기
fe_df = pd.read_csv('data/6_feature_engineering.csv')
fe_df.date = pd.to_datetime(fe_df.date)
fe_df = fe_df.drop(['전일 주가','주가하락세', '현재지수', '등락률(%)'], axis=1)
fe_df = fe_df.bfill()

fe_df = fe_df.set_index('date')
dfx = fe_df[list(fe_df.iloc[:, 1:].columns)]
dfy = fe_df[['price']]

X = dfx.values.tolist()
y = dfy.values.tolist()

print(X[1],y[1])
```

```python
window_size = 10

data_X = []
data_y = []
for i in range(len(y) - window_size):
    _X = X[i : i + window_size] # 다음 날 종가(i+windows_size)는 포함되지 않음
    _y = y[i + window_size]     # 다음 날 종가
    data_X.append(_X)
    data_y.append(_y)

print('전체 데이터의 크기 :', len(data_X), len(data_y))
```

```python
train_size = int(len(data_y) * 0.7)
train_X = np.array(data_X[0 : train_size])
train_y = np.array(data_y[0 : train_size])

test_size = len(data_y) - train_size
test_X = np.array(data_X[train_size : len(data_X)])
test_y = np.array(data_y[train_size : len(data_y)])

print('훈련 데이터의 크기 :', train_X.shape, train_y.shape)
print('테스트 데이터의 크기 :', test_X.shape, test_y.shape)
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout

model = Sequential()
model.add(LSTM(units=20, activation='relu', return_sequences=True, input_shape=(10, 119)))
model.add(Dropout(0.1))
model.add(LSTM(units=20, activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(units=1))
model.summary()
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(train_X, train_y, epochs=70, batch_size=30)
pred_y = model.predict(test_X)
```

```python
pred_y = model.predict(test_X)

plt.figure()
plt.plot(test_y, color='red', label='주가')
plt.plot(pred_y, color='blue', label='10일간격 예측 주가')
plt.title('오아시스 주가 예측')
plt.xlabel('date')
plt.ylabel('주가')
plt.legend()
plt.show()

print("22/10/1 주가 :", fe_df.price[-1] * pred_y[-1] / dfy.price[-1])
```
