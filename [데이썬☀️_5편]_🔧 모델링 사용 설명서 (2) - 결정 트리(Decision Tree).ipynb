## **1. 분류의 개요**

### **1) 분류란?**

분류는 지도학습의 대표적인 유형입니다. 

지도학습은 명시적인 정답이 있는 데이터가 주어진 상태에서 학습하는 머신러닝 방법입니다.


분류에서는 기존 데이터가 어떤 레이블에 속하는지 패턴을 알고리즘으로 인지한 뒤

새롭게 관측된 데이터에 대한 레이블을 판별합니다.

<br>

**[분류의 머신러닝 알고리즘]**

- **나이브 베이즈** : 베이즈 통계와 생성 모델
- **로지스틱 회귀** : 독립변수와 종속변수의 선형 관계성
- **결정 트리** : 데이터 균일도에 따른 규칙
- **서포트 백터 머신** : 개별 클래스 간의 최대 분류 마진을 효과적으로 찾아 줌
- **최소 근접** : 근접 거리 기준
- **신경망** : 심층 연결
- **앙상블** : 서로 다른 또는 같은 머신러닝 알고리즘 결합

<br>

## **2. 결정 트리🌳**

### **1) 결정 트리란?** 

결정 트리란 데이터에 있는 규칙을 학습을 통해 자동으로 찾아내

트리(Tree) 기반의 분류 규칙을 만드는 것 입니다. 

어떤 기준을 기반으로 규칙을 생성하는지가 알고리즘의 성능과 효율성을 좌지우지 합니다.

결정트리.jpg

- **규칙노드(Decision Node)** : 규칙조건
- **리프노드(Leaf Node)** : 결정된 클래스 값
- **서브 트리(Sub Tree)** : 새로운 규칙 조건 

<br>

규칙이 많아질 수록 즉, 트리의 깊이가 깊어질수록 결정 트리의 예측력은 낮아질 가능성이 높습니다.

따라서 트리의 분할은 최대한 균일한 데이터 세트를 구성할 수 있도록 분할해야 합니다.

<br>

아래 그림에서 균일도가 가장 높은 데이터 세트는 어느 것 일까요?😀

균일도.jpg


정답은 'B' 입니다.

B에서 하나의 데이터를 선정했을 때 데이터에 대한 구체적인 정보없이도 '검정 공⚫'이라고 쉽게 예측할 수 있습니다. 

반면, C의 경우는 상대적으로 혼잡도가 높고 균일도가 낮기 때문에 같은 조건에서 데이터를 예측하는 데는 더 많은 정보가 필요합니다.

<br>

**[결정 노드]**

결정 노드는 정보 균일도가 높은 데이터 세트를 먼저 선택할 수 있는 규칙 조건을 만듭니다. 

정보 균일도가 데이터 세트로 분할될 수 있도록 조건을 찾아 서브 데이터 세트를 만들고,


다시 이 서브 데이터 세트에서 균일도가 높은 자식 데이터 세트를 분활하는 방식을 반복하여 데이터 값을 예측합니다.

위의 그림에서 가장 첫 번째로 만들어져야 하는 규칙 조건은 무엇일까요?😁

레고.jpg



정답은 '녹색' 입니다. 모두 녹색 세모로 가장 예측하기 쉽기 때문입니다. 


<br>

**[정보의 균일도를 측정하는 방법]**

- **정보 이득** : '(1 - 엔트로피 지수)'로 정보 이득이 높은 속성을 기준으로 분할


> 엔트로피 지수 : 데이터 집합의 혼잡도

<br>

- **지니 계수** : 지니 계수가 낮은 속성을 기준으로 분할


> 지니계수 : 경제학에서 불평등 지수를 나타낼 때 사용하는 계수

### 2) **결정 트리 모델의 특징**

**[결정 트리의 장점]**

결정 트리의 가장 큰 장점은 알고리즘이 쉽고 직관적이라는 점입니다. 

'균일도'라는 룰이 매우 명확하여 각 피처의 스케일링과 정규화 같은 전처리 작업이 필요없으며

이를 기반으로 규칙 노드와 리프 노드가 어떻게 만들어 지는지를 알 수 있고 시각화로 표현이 가능하기 때문입니다. 

**[결정 트리의 단점]**

가장 큰 단점은 과적합으로 정확도가 떨어질 수 있다는 점입니다. 

학습 데이터 기반 모델의 정확도를 높이기 위해 계속해서 조건을 추가하면서 깊이가 점점 깊어집니다.

복잡한 학습 모델은 새로운 데이터 세트가 들어 왔을 때 예측 성능이 떨어질 수 밖에 없습니다. 

따라서, 트리의 크기를 사전에 제한하는 것이 성능 튜닝에 도움이 됩니다.

<br>

### 3) **결정 트리 파라미터**

결정 트리 알고리즘은 사이킷런의 DecisionTreeClassifier와 DecisionTreeRegressor 로 구현할 수 있습니다. 

DecisionTreeClassifier는 분류를 위한 클래스 이고 DecisionTreeRegressor는 회귀를 위한 클래스 입니다. 



<BR>

- **min_samples_split** : 자식 노드를 분활하기 위한 최소한의 샘플 데이터 수, 디폴트 값은 2


- **min_samples_leaf** : 말단 노드(leaf)에 있어야 할 최소한의 샘플 데이터 수, 디폴트 값은 1


- **max_features** : 각 노드에서 분활에 사용할 변수의 최대 수
  * sqrt : 전체 변수 개수의 제곱근 
  * auto : sqrt와 동일
  * log : 전체 변수 중 log2(전체 변수 개수) 선정
  * None : 전체 변수 선정



- **max_depth**: 트리의 최대 깊이, 디폴트 값은 None
  * None : 완벽하게 클래스 결정 값이 될 때 까지 혹은 데이터 개수가 min_samples_split보다 작아질 때까지 계속 깊이 증가 

  

- **max_leaf_nodes** : 잎 노드의 최대 개수, 디폴트 값은 None

색깔이 진할수록 해당 레이블에 속하는 샘플 데이터 건수가 많습니다.

### **5) 파라미터 변경**


결정 트리는 규칙 생성 로직을 미리 제어하지 않으면 완벽한 클래스 값을 구별해낼 때 까지 트리 노드를 계속해서 만들어 갑니다.

이로 인해 복잡한 규칙 트리가 생성되어 과적합이 발생할 수 있기 때문에 파라미터를 변경하며 알고리즘을 제어해줍니다.

결정 트리 파라미터 중 max_depth, min_samples_split, min_samples_leaf을 변경해보며 각각의 트리의 차이점을 살펴보겠습니다.

<br>

#### **[max_depth]**

max_depth는 나무의 최대 깊이를 제어하고 디폴트 값은 None입니다. 

max_depth를 4로 변경해보겠습니다.

```python
tree = DecisionTreeClassifier(max_depth=4, random_state=0)
tree.fit(X_train, y_train)
```

#### **[min_samples_split]**

기존 트리의 min_sampels_split은 디폴트 값인 2를 가지고 있었습니다.

이번에는 min_samples_split을 20으로 제어해보겠습니다.

나무의 깊이도 얕아지고 리프 노드의 샘플 개수도 커졌음을 살펴볼 수 있습니다.

#### **[min_samples_leaf]**

min_samples_leaf는 리프 노드가 될 수 있는 샘플 데이터 건수의 최솟값을 지정합니다. 디폴트 값은 1입니다.

### **7) 모델의 정확도 측정**

파라미터 변경해보며 모델의 정확도가 어떻게 달라지는지 측정할 수 있습니다.

```python
tree = DecisionTreeClassifier(max_depth=4, min_samples_split=20, min_samples_leaf=12, random_state=0)
tree.fit(X_train, y_train)

print("학습용 데이터 정확도: {:.3f}".format(tree.score(X_train, y_train)))
print("시험용 데이터 정확도: {:.3f}".format(tree.score(X_test, y_test)))
```

앞서 변경했던 파라미터 값들을 적용하여 모델의 정확도를 측정해보았습니다.

학습용 데이터의 정확도는 0.948이고 시험용 데이터 정확도는 0.958이 나왔습니다.
