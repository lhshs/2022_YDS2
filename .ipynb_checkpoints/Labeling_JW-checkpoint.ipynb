## 2-1. 날짜 형식 어긋난 것 가려내서 전부 nan 처리하고 ffill 함

```python
# df 에서 날짜가 이상한 애들만 골라서 보기
df[df["날짜"].str.contains('199|200|201|202')==False]
```

```python
# 위의 이상한 애들만 골라서 날짜를 np.nan 으로 뱌꾸기
df.loc[df["날짜"].str.contains('199|200|201|202')==False, "날짜"] = np.nan
```

```python
# 이상한 애들 다시 확인하기 => 하나도 보이지 않음
df[df["날짜"].str.contains('199|200|201|202')==False]
```

```python
# isnull() 로 제대로 바뀌었는지 확인 => 행의 개수가 이상한 애들이었을 때의 204576 rows로 같음 위위위 줄에서 확인 가능
df[df.isnull()["날짜"]]
```

```python
# ffill 로 nan 값을 전부 앞의 값으로 치환함
df.fillna(method='ffill', inplace=True)
```

```python
# 더 이상 nan 값이 없음을 확인함
df[df.isnull()["날짜"]]
```

## 2-2. 날짜 "." 전부 "-" 으로 치환, 끝에 "-" 은 제거하기

```python
df["날짜"] = df["날짜"].str.replace(".", "-").str.rstrip("-")
df
```

## 3. df 날짜 fdr 삼성전자 데이터랑 맞춰서 자름

```python
# 삼성전자 데이터 시작일이 1998-05-25 이지만 26일부터 Change 값이 기록돼있으므로 df 의 "날짜" 열의 시작일을 25일로 맞춰줌
df = df[df["날짜"] >= "1998-05-25"]
df
```

```python
# 인덱스를 맞춰줌
df.reset_index(drop=True, inplace=True)
df
```

## 4-1. Change 열 생성후 날짜를 하루씩 뒤로 미뤄가면서 fdr 삼성전자 데이터와 비교해서 찾아감

```python
df["Change"] = 0
df
```

## 4-2. 찾아지면 그 행의 Change > 0 이면, +1 < 0 이면 -1 을 df.Change 열에 대입함

```python
# transform 함수를 적용하기 위해 함수를 만듦
# df["날짜"] 열을 하나씩 입력으로 받아서 df["Change"]의 값을 상승이면 +1 하락이면 -1로 바꾸기로 함
def change_of_next_day_of_stock_market(today):
    return_change = 0
    today = datetime.date.isoformat(datetime.date.fromisoformat(today) + datetime.timedelta(days=1)) # 하루 미룸
    while return_change == 0:
        try:
            if stock_df.loc[today]["Change"] > 0:
                return_change = 1 # 찾은 "Change" 열의 값이 > 0 이면 상승이니까 + 1을 반환
            else:
                return_change = -1 # 찾은 "Change" 열의 값이 < 0 이면 하락이니까 - 1을 반환
        except:
            today = datetime.date.isoformat(datetime.date.fromisoformat(today) + datetime.timedelta(days=1))
            # 삼성전자 데이터에서 날짜가 찾아지질 않으면 하루 더 미룸 while 때문에 찾아질 때까지 뒤로 하루씩 미룸
    return return_change
```

```python
# "Change" 열을 바꿈, 입력열은 "날짜" 열이고 적용함수는 바로 위 셀에 정의된 change_of_next_day_of_stock_market)
# 꽤 오랜 시간이 걸리는 셀이므로 확인은 다음 셀에서 함
df["Change"] = df["날짜"].transform(change_of_next_day_of_stock_market)
```
