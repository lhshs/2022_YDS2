```python
import pandas as pd
import numpy as np
import FinanceDataReader as fdr

from tqdm.notebook import tqdm
import time

from datetime import datetime, timedelta
```

```python
df = pd.read_csv('D:/Documents/GitHub/NLP_pjt/df_final_Date_refine_add_blank.csv', index_col = 0).reset_index().drop(columns='index')
df = df.rename(columns = {'언론사':'Press', '날짜':'Date', '제목':'Title'})
stock_df = fdr.DataReader('005930')

df = df[df['Date'] >= str(stock_df.index[1])[:10]].reset_index().drop(columns='index')
df = df.groupby(['Date', 'Press']).sum()

```

```python
df['Change'] = 'x'
```

```python
df
```

```python
stock_df[1:50]
```

```python
j = 0
n = 0

for i in tqdm(stock_df.index[1:50]): 
    date_list=[]
    temp_list = []
    while True:
        j += 1
        stock = datetime.strptime(str(i), "%Y-%m-%d %H:%M:%S") # timedelta(days=1) # datetime.datetime(1998, 5, 29, 0, 0) 형태로 변환
        news = datetime.strptime(df.index[j-1][0], "%Y-%m-%d") # + timedelta(days=1)
        temp_list.insert(0, df.index[j-1][0])
        date_list = temp_list
        
        if stock > news:
            news = news + timedelta(days=1) # 휴일장, 주식 날짜 이하로 뉴스 모아서 label
            continue
        elif stock < news: # 장은 열렸지만, news는 없을 때
            j -= 1 # 보류, 날짜 유지, 뉴스기사 버리지 말아라
            stock = stock - timedelta(days=1)
            print('다음 날의 주가 데이터')
            break
        elif news == stock:
            break
        
    while n <= len(date_list):
        if stock_df.loc[str(i), 'Change'] < 0:
            df.loc[df.index.get_level_values('Date').str.contains(date_list[n]), 'Change'] = -1
            break
        elif stock_df.loc[str(i), 'Change'] > 0:
            df.loc[df.index.get_level_values('Date').str.contains(date_list[n]), 'Change'] = 1
            break
        else:
            df.loc[df.index.get_level_values('Date').str.contains(date_list[n]), 'Change'] = 0
            break
        
    print(date_list)
    print(len(date_list))
    print('가져와야 할 주가 데이터 : ', stock)
    print("----------")
```
